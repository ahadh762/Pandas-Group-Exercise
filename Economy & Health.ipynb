```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_excel('Inflation Percentage.xlsx')
df.drop([0,1,2], axis=0, inplace=True)
df.drop('_63',axis=1,inplace=True)
year = 1960
for i in range (2,63):
    df=df.rename(columns={f"_{i}":f"{year}"})
    year+=1
df = df[(df['Data Source'] == 'United States') | (df['Data Source'] == 'World')]
df = df.T

df.drop(['Data Source','World Development Indicators','Column1','_1'], axis=0, inplace=True)
df.rename(columns={255:"United States",263:"World"},inplace=True)
df

ax = df.plot(title = 'Inflation Percentage by Year',kind='line')
```
