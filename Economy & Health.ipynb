```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_excel('Inflation Percentage.xlsx')
df.drop([0,1,2], axis=0, inplace=True)
df.drop('_63',axis=1,inplace=True)
year = 1960
for i in range (2,63):
    df=df.rename(columns={f"_{i}":f"{year}"})
    year+=1

template = df.copy()  # Make a copy of the dataframe with all the changes I made so I don't have to do it again

df = df[(df['Data Source'] == 'United States') | (df['Data Source'] == 'World')]
df = df.T

df.drop(['Data Source','World Development Indicators','Column1','_1'], axis=0, inplace=True)
df.rename(columns={255:"United States",263:"World"},inplace=True)
df

ax = df.plot(title = 'Inflation % by Year (USA vs. World)',kind='line')
ax.set_xlabel('Year')
ax.set_ylabel('Inflation %')
plt.show()
```

```python
df = template.copy()
df = df[(df['Data Source'] == 'United States') | (df['Data Source'] == 'China') | (df['Data Source'] == 'European Union')]
df = df.T

df.drop(['Data Source','World Development Indicators','Column1','_1'], axis=0, inplace=True)

df.rename(columns={255:"United States",44:"China", 77:"European Union"},inplace=True)

ax = df.plot(title = 'Inflation % by Year (Global Super Powers)', kind='line')
ax.set_xlabel('Year')
ax.set_ylabel('Inflation %')
plt.show()
```
